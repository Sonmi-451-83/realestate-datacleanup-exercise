```python
from ipyleaflet import Map, basemaps, Polyline, CircleMarker

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (40.4168, -3.7038), zoom = 10, min_zoom = 1, max_zoom = 20, basemap=basemaps.OpenStreetMap.Mapnik)

# basemap=basemaps.Stamen.Terrain
map
```

```python
dict_location = []

for index, file in ds3.iterrows():
     dict_location.append([file["level5"], file["latitude"], file["longitude"]])

for loc in dict_location:
     loc[1] = float(loc[1].replace(',', '.'))

for loc in dict_location:
     loc[2] = float(loc[2].replace(',', '.'))


Fuenlabrada_layer = []
Getafe_layer = []
Alcorcón_layer = []
Leganés_layer = []

for item in dict_location:
     if item[0] == "Fuenlabrada":
          Fuenlabrada_layer.append([item[1], item[2]])
     elif item[0] == "Getafe":
          Getafe_layer.append([item[1], item[2]])
     elif item[0] == "Alcorcón":
          Alcorcón_layer.append([item[1], item[2]])
     elif item[0] == "Leganés":
          Leganés_layer.append([item[1], item[2]])

line_f = Polyline(locations=Fuenlabrada_layer, color="blue", weight=4)
line_g = Polyline(locations=Getafe_layer, color="red", weight=4)
line_a = Polyline(locations=Alcorcón_layer, color="green", weight=4)
line_l = Polyline(locations=Leganés_layer, color="yellow", weight=4)

#map.add(line_f)

for coord in Fuenlabrada_layer:
    circle_marker = CircleMarker()
    circle_marker.location = coord
    circle_marker.radius = 2
    circle_marker.color = "blue"
    circle_marker.fill_color = "blue"
    map.add(circle_marker)


#map.add(line_g)

for coord in Getafe_layer:
    circle_marker = CircleMarker()
    circle_marker.location = coord
    circle_marker.radius = 2
    circle_marker.color = "red"
    circle_marker.fill_color = "red"
    map.add(circle_marker)

#map.add(line_a)

for coord in Alcorcón_layer:
    circle_marker = CircleMarker()
    circle_marker.location = coord
    circle_marker.radius = 2
    circle_marker.color = "green"
    circle_marker.fill_color = "green"
    map.add(circle_marker)

#map.add(line_l)

for coord in Leganés_layer:
    circle_marker = CircleMarker()
    circle_marker.location = coord
    circle_marker.radius = 2
    circle_marker.color = "yellow"
    circle_marker.fill_color = "yellow"
    map.add(circle_marker)

map


```
